```python
# 로지스틱, dt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score

from imblearn.combine import SMOTEENN
import xgboost
import lightgbm
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.3, random_state = 42,
                                                   stratify=y)
```

```python
X_resampled, y_resampled = SMOTEENN(random_state=42, sampling_strategy=0.7).fit_resample(X_train,y_train)
```

```python
y_resampled.value_counts()
```

# GBT

# gridSearch

```python
def clf3():
    gbt = GradientBoostingClassifier()
    xgb = xgboost.XGBClassifier() 
    lgbm = lightgbm.LGBMClassifier()
  
    return gbt, xgb, lgbm
```

```python
gbt_parameters ={'max_depth' : [6,8],
                 'n_estimators': [10,20,30],
                 'learning_rate':[0.05,0.1]}

xgb_parameters ={'max_depth' : [3,4,5,6,7],
                'n_estimators': [50,100,150],
                'learning_rate':[0.01,0.05,0.1,0.5],
                'gamma': [0.5, 1, 2]}

lgbm_parameters ={'boosting_type' : ['gbdt','goss'], 
                'learning_rate': [0.05,0.1,0.2],
                'n_estimators' : [100, 200, 300]}
```

```python
from sklearn import model_selection
gbt, xgb, lgbm = clf3()
#GBT, scoring기준은 accuracy값으로, cross-validation은 10회 반복한다는 뜻임
grid_search_gbt = model_selection.GridSearchCV ( estimator = gbt,
                                                param_grid = gbt_parameters,
                                                scoring = 'accuracy', cv = 3 )
 
grid_search_gbt.fit(X_resampled, y_resampled)
#best hyper parameter값을 받음
best_gbt = grid_search_gbt.best_estimator_


# XGB
grid_search_xgb = model_selection.GridSearchCV ( estimator = xgb,
                                                param_grid = xgb_parameters,
                                                scoring = 'accuracy', cv = 3 )
 
grid_search_xgb.fit(X_resampled, y_resampled)
best_xgb = grid_search_xgb.best_estimator_

# LGBM
grid_search_lgbm = model_selection.GridSearchCV ( estimator = lgbm,
                                                 param_grid = lgbm_parameters,
                                                 scoring = 'accuracy', cv = 3 )
 
grid_search_lgbm.fit(X_resampled, y_resampled)
best_lgbm = grid_search_lgbm.best_estimator_
```
